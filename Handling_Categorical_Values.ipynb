```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# ==========================================
# 1. GENERATING MOCK DATA WITH CATEGORICAL COLUMNS
# (Replace this with your actual dataset load)
# ==========================================
np.random.seed(42)
n_samples = 5

data = {
    'Age': [34, 45, 22, 51, 29],
    'Income': [75000, 42000, 120000, 95000, 30000],
    # Categorical columns that need to be transformed:
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Education_Level': ['Graduate', 'High School', 'Postgraduate', 'Graduate', 'High School'],
    'Prior_Default': ['No', 'Yes', 'No', 'No', 'No'],
    'Approved': [1, 0, 1, 1, 0]
}

df = pd.DataFrame(data)

print("--- Original Dataset (Before Label Encoding) ---")
print(df)
print("\n" + "="*50 + "\n")

# ==========================================
# 2. DEFINING THE CATEGORICAL HANDLING FUNCTION
# ==========================================
def handle_categorical_values(dataset):
    """
    Uses LabelEncoder to convert categorical/text columns into
    integer values using the fit_transform() method.
    """
    # Create a copy of the dataframe to avoid modifying the original data
    processed_df = dataset.copy()

    # Initialize the LabelEncoder class
    label_encoder = LabelEncoder()

    # Identify columns that contain textual/categorical data (object type)
    categorical_columns = processed_df.select_dtypes(include=['object']).columns

    print(f"Categorical columns found to encode: {list(categorical_columns)}\n")

    # Loop through each categorical column and transform it
    for col in categorical_columns:
        # fit_transform identifies unique category names and converts them to integers
        processed_df[col] = label_encoder.fit_transform(processed_df[col])

    return processed_df

# ==========================================
# 3. EXECUTING THE PROCESS
# ==========================================
# Run the function on our dataset
encoded_df = handle_categorical_values(df)

print("--- Processed Dataset (After Label Encoding) ---")
print(encoded_df)
print("\n> Note: Text labels are now numeric representations, ready for Tree-based models.")
```
